```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def buy_call_payoff(S, K, P):
    return np.maximum(0, S - K) - P

def sell_call_payoff(S, K, P):
    return P - np.maximum(0, S - K)

def buy_put_payoff(S, K, P):
    return np.maximum(0, K - S) - P

def sell_put_payoff(S, K, P):
    return P - np.maximum(0, K - S)
```

```python
K = 100  # Strike Price
P = 10   # Premium
S = np.arange(K - 30, K + 31, 1)  # Spot Price range from K-50 to K+50
```

```python
def plot_option_payoffs(S, K, P):
    payoffs = [
        (buy_call_payoff(S, K, P), 'Buy Call', 'blue'),
        (sell_call_payoff(S, K, P), 'Sell Call', 'red'),
        (buy_put_payoff(S, K, P), 'Buy Put', 'green'),
        (sell_put_payoff(S, K, P), 'Sell Put', 'orange')
    ]

    fig, axs = plt.subplots(2, 2, figsize=(12, 8))
    fig.suptitle('Option Payoff Diagrams')

    for ax, (payoff, title, color) in zip(axs.flatten(), payoffs):
        ax.plot(S, payoff, color=color)
        ax.axhline(0, color='black', linewidth=0.5)
        ax.set_title(title)
        ax.set_xlabel('Spot Price (S)')
        ax.set_ylabel('Profit / Loss')
        ax.grid(True)

    plt.tight_layout(rect=[0, 0.03, 1, 0.95])
    plt.show()
```

```python
plot_option_payoffs(S, K, P)
```

### Observations

- **Buy Call:** Loss is limited to the premium; profit increases if spot price rises above strike + premium.
- **Sell Call:** Maximum profit is the premium; losses increase as spot price rises above strike.
- **Buy Put:** Loss is limited to the premium; profit increases as spot price falls below strike - premium.
- **Sell Put:** Maximum profit is the premium; losses increase as spot price falls below strike.
